```python

# Write a SQL query to identify customers (Lastname and Firstname) from California in chinook.db (20 points)
query_1 = """
SELECT LastName, FirstName
FROM customers
WHERE State = 'CA';
"""

df1 = pd.read_sql_query(query_1, conn)
display(df1)
```

```python
#  Write a SQL query to identify customers whose lastname initial is “G” (20 points)
query_2 = """
SELECT CustomerId, LastName, FirstName
FROM customers
WHERE LastName LIKE 'G%';
"""

df2 = pd.read_sql_query(query_2, conn)
display(df2)
```

```python
# List the “Tracks” (name) purchased by customers identified in step 2. (20 points)
query_3 = """
SELECT DISTINCT tracks.Name
FROM invoice_items
JOIN invoices ON invoice_items.InvoiceId = invoices.InvoiceId
JOIN customers ON invoices.CustomerId = customers.CustomerId
JOIN tracks ON invoice_items.TrackId = tracks.TrackId
WHERE customers.LastName LIKE 'G%';
"""

df3 = pd.read_sql_query(query_3, conn)
display(df3)
```

```python
# List the “Artists” (name) for the albums identified in step 3. (20 points)
query_4 = """
SELECT DISTINCT artists.Name
FROM artists
JOIN albums ON artists.ArtistId = albums.ArtistId
JOIN tracks ON albums.AlbumId = tracks.AlbumId
WHERE tracks.Name IN (
    SELECT DISTINCT tracks.Name
    FROM invoice_items
    JOIN invoices ON invoice_items.InvoiceId = invoices.InvoiceId
    JOIN customers ON invoices.CustomerId = customers.CustomerId
    JOIN tracks ON invoice_items.TrackId = tracks.TrackId
    WHERE customers.LastName LIKE 'G%'
);
"""

df4 = pd.read_sql_query(query_4, conn)
display(df4)
```

```python
# Sort the Artist name list in DESC order. (20 points)
query_5 = """
SELECT DISTINCT artists.Name
FROM artists
JOIN albums ON artists.ArtistId = albums.ArtistId
JOIN tracks ON albums.AlbumId = tracks.AlbumId
WHERE tracks.Name IN (
    SELECT DISTINCT tracks.Name
    FROM invoice_items
    JOIN invoices ON invoice_items.InvoiceId = invoices.InvoiceId
    JOIN customers ON invoices.CustomerId = customers.CustomerId
    JOIN tracks ON invoice_items.TrackId = tracks.TrackId
    WHERE customers.LastName LIKE 'G%'
)
ORDER BY artists.Name DESC;
"""

df5 = pd.read_sql_query(query_5, conn)
display(df5)
```
